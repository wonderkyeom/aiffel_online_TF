```python
import math

stream = [] # 스트림스
cnt = 0
sum = 0
score = {'1':0, '2':1, '3':3, '4':5, '5':7, '6':9, '7':11, '8':15, '9':20, '10':25, '11':30, '12':35, '13':40, '14':50, '15':60, '16':70, '17':85, '18':100, '19':150, '20':300}

while cnt< 20:
    num = int(input("뽑은 숫자를 말씀해주세요!"))
    cnt+=1

    stream.append(num)
    print(stream)

#############################점수산정

asc = 1 #오름차순 카운트
end = False
for i in range(len(stream) - 1):
    print(i, asc)

    # 안짤리면 계속 asc 누적
    if stream[i] <= stream[i+1]:
        end = False
        asc += 1
    else:
        end = True
        sum += score.get(str(asc)) # asc 키에 해당하는 value 더하기
        asc = 1

    if end: # 짤리면
#        print("합침", asc)
#        sum += score.get(str(asc)) # asc 키에 해당하는 value 더하기
#        print(sum)
        end = False

sum += score.get(str(asc)) # asc 키에 해당하는 value 더하기

print("점수 : ", sum)





'''
def streams():
    if len(num) > 20:
        print("숫자는 20개 내로 뽑아주세요!")
        return False

    for i in num:
        stream.append(i)

    # stream을 순차적으로 돌면서 오름차순일때 까지 sum을 누적

    # 한번이라도 오름차순이 끊기면 다시 0으로 초기화
    for i in range(len(stream)-1):
      if stream[i]< stream[i+1]:
        sum += 1


print(streams()) # 최종 합산 점수
print(stream)
'''
```
